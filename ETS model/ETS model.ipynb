# ETS model - Decomposition of time series data

### Importing package
The required packages are imported. These are pandas for working with data and matplotlib for plotting graphs.

```python
import pandas as pd
import matplotlib.pyplot as plt
%matplotlib inline
```

### Analyze data
The dataset used here is the sales data. It is a univariate time series data meaning that it has only one variable. The index column is assigned as the Month column of the dataset.

```python
df = pd.read_csv('sales.csv',index_col='Month')
```

```python
df.head()
```

```python
df.plot()
```

To be able to work with the data, the 'month' colume is changed to a datetime index form.

```python
df.index = pd.to_datetime(df.index)
df.head()
```

### Decomposition
The decomposition will be done using the statsmodels library. It has a seasonal_decompose function. This function takes the column name for which the season has to be decomposed and the type of seasonal component. It can be additive or multiplicative. It takes hit and trial to see which component to use. By seeing the graph we can see that it is not linear therefore the type has to be multiplicative.

```python
from statsmodels.tsa.seasonal import seasonal_decompose
```

```python
final = seasonal_decompose(df['Sales'],model='multiplicative')
```

Now from this you can grab the seasonal component and the trend or even plot all together.

```python
final.trend.plot()
```

As seen from the above plot, the trend is upwards.

```python
final.seasonal.plot()
```

```python
fig = final.plot()
```
